```python
# Install the necessary libraries from the requirements.txt file into your corresponding virtual environment. Make sure you start JupyterLab with the extra data rate argument.
from textblob import TextBlob
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt  # Corrected typo
import nltk
import re
from collections import Counter

# Download necessary NLTK data
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

sns.set()  # Set Seaborn styles
```

```python
# Loading the content of the file (Load the twentieth-century data that you scraped in Exercise 1.4.)
file_path = '/Users/renubalaji/Documents/GitHubProjects/20th-century/key_events_20th_century.txt'

with open(file_path, 'r') as file:
    data = file.read()

# Print a snippet of the content
print(data[:1000])  # print the first 1000 characters for preview
```

```python
# Manually tokenizing the text using regular expressions (Tokenize the words from the text and create a bar chart to plot the 10 most common words.)
tokens = re.findall(r'\b\w+\b', data.lower())
```

```python
# Counting the frequency of each word
word_freq = Counter(tokens)

# Getting the 10 most common words
common_words = word_freq.most_common(10)

# Splitting the words and their frequencies for plotting
words, counts = zip(*common_words)

# Creating a bar chart for the 10 most common words
plt.figure(figsize=(10, 6))
plt.bar(words, counts, color='skyblue')
plt.title("Top 10 Most Common Words in 20th Century Events")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Import necessary libraries
import re
from collections import Counter
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
```

```python
# Load stop words from NLTK 
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

```python
# Remove stop words and non-alphabetic tokens (Remove stop words (and punctuation marks if necessary), and rerun the plot. What has changed? Comment on your impressions in a markdown cell under the plot.)
filtered_tokens = [word for word in tokens if word.isalpha() and word not in stop_words]

# Count the frequency of each word
word_freq = Counter(filtered_tokens)

# Get the 10 most common words
common_words = word_freq.most_common(10)

# Split the words and their frequencies for plotting
words, counts = zip(*common_words)

# Create a bar chart for the 10 most common words
plt.figure(figsize=(10, 6))
plt.bar(words, counts, color='skyblue')
plt.title("Top 10 Most Common Words (Stop Words Removed)")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

#After removing the stopping words, the words that don't contribute any sense to the analysis are all removed and words which make sense are retained.
```

```python
# Download the required NLTK resources (Create a tags list using TextBlob)

from textblob import TextBlob

# Create a TextBlob object
blob = TextBlob(data)

# Generate a list of (word, tag) tuples using TextBlob's part-of-speech tagger
tags = blob.tags

# Display the first 10 tags as an example
print(tags[:10])
```

```python
# POS Tagging (Create a list of the Top 10 POS tags for words that appear in the article)
sans_punc = re.sub("[^a-zA-Z]",  # Search for all non-letters
                        " ",        # Replace all non-letters with spaces
                        str(filtered_tokens))
```

```python
# Word tokenization
from nltk.tokenize import word_tokenize
tokenized_word_2 = word_tokenize(sans_punc)
print (tokenized_word_2)
```

```python
from nltk.probability import FreqDist
dist_words = FreqDist(tokenized_word_2)
print(dist_words)
```

```python
#Create a plot showing the frequency of the number of times these countries are mentioned in the text. Is there anything in particular you notice in this plot?
# Plotting the frequency of mentions
plt.figure(figsize=(10, 5))
plt.bar(df_countries['Country'], df_countries['Mentions'], color='skyblue')
plt.title('Frequency of Country Mentions in the Text')
plt.xlabel('Country')
plt.ylabel('Number of Mentions')
plt.xticks(rotation=45)
plt.grid(axis='y')

# Show the plot
plt.show()
```
